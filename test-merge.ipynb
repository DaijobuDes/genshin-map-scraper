```python
from PIL import Image
from matplotlib.pyplot import imshow
from enums import *
import math
import os
import numpy as np
```

```python
# Specify width x height
obj = GOLDEN_APPLE_ARCHIPELAGO_13

w = (abs(obj.getXmin()) + abs(obj.getXmax())+1) * 256
h = (abs(obj.getYmin()) + abs(obj.getYmax())+1) * 256

print([w, h])
```

```python
directory = os.listdir("map-isles\\GOLDEN_APPLE_ARCHIPELAGO_13")

print(f"Found {len(directory)} files")
```

```python
# Initialize new white image
background = Image.new("RGBA", (w, h), (255, 255, 255, 255))
# imshow(np.asarray(background))
```

```python
x = 0
y = 0
for i in range(obj.getXmin(), obj.getXmax()+1):
    for j in reversed(range(obj.getYmin(), obj.getYmax()+1)):
        # print(f"tile-{i}_{j}.jpg")
        tile = Image.open(f"map-isles\\GOLDEN_APPLE_ARCHIPELAGO_13\\tile-{i}_{j}.jpg")
        background.paste(tile, (x*256, y*256))
        y += 1
    y = 0
    x += 1

# display(background)
imshow(np.asarray(background))
background.save("image.png")
```
